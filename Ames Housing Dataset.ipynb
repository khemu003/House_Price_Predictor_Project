```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AmesHousing.csv")
df.head()
```

# For Handling too much Null values columns

```python
missing_value_percent = (df.isnull().sum() / len(df) * 100).reset_index(name="count")
missing_value_percent
```

```python
x = missing_value_percent[missing_value_percent["count"] > 50]
df = df.drop(columns = x["index"])
```

```python
df.shape
```

# Filter numerical and categorical data type in different variable

```python
numerical_col = df.select_dtypes(include=["int64", "float64"]).columns
categorical_col = df.select_dtypes(include=["object"]).columns

print("Numerical columns : ",numerical_col)

print("Categorical columns : ",categorical_col)
```

# Handling Missing value

```python
from sklearn.impute import SimpleImputer

num_si = SimpleImputer(strategy="mean")
cat_si = SimpleImputer(strategy="most_frequent")

df[numerical_col] = num_si.fit_transform(df[numerical_col])
df[categorical_col] = cat_si.fit_transform(df[categorical_col])
```

```python
df.info()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in categorical_col:
    df[i] = le.fit_transform(df[i])

df.head(3)
```

```python
important_features = [
    'Overall Qual', 'Overall Cond', 'Gr Liv Area', 'Total Bsmt SF', '1st Flr SF',
    'Garage Area', 'Lot Area', 'Year Built', 'Year Remod/Add', 'Neighborhood',
    'Bsmt Qual', 'BsmtFin SF 1', 'Full Bath', 'Bsmt Full Bath', 'Kitchen Qual',
    'Garage Cars', 'Wood Deck SF', 'Fireplaces'
]
```

# train_test_split

```python
from sklearn.model_selection import train_test_split

x = df[important_features]
y = df["SalePrice"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train_sc = sc.fit_transform(x_train)

x_train_sc_df = pd.DataFrame(x_train_sc, columns = x_train.columns)

np.round(x_train_sc_df.describe(), 2)
```

# MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mmc = MinMaxScaler()

x_train_mmc = mmc.fit_transform(x_train)

x_train_mmc_df = pd.DataFrame(x_train_mmc, columns = x_train.columns)

np.round(x_train_mmc_df.describe(), 2)
```

# Apply LinearRegression Model because target column is numerical

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

score = r2_score(y_test, y_pred)
score
```
